# EllipticalEnvelope method

```python
# Scaling the data
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler() 
df1_ss = ss.fit_transform(df1_num)
```

```python
from sklearn.covariance import EllipticEnvelope
```

```python
# Applying the Ellipticenvelope method
```

```python
clf = EllipticEnvelope(contamination=.1,random_state=0)
clf.fit(df1_ss)
```

```python
ee_scores = pd.Series(clf.decision_function(df1_ss))  #Returns the mean accuracy on the given da
ee_predict = clf.predict(df1_ss) #Predict the labels (1 inlier, -1 outlier) of X according to the fitted model.
```

```python
ee_scores 
```

```python
pd.Series(ee_predict).value_counts() # Lables with -1 are the outliers
```

```python
df1['Outliers']=pd.DataFrame(ee_predict)
```

```python
outlier_data=df1[df1['Outliers']==-1].reset_index()
```
